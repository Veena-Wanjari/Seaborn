# What is Seaborn??

* Seaborn is a Python Data Visualization library.

* The seaborn package was developed based on the Matplotlib library. 

* It is used to create more attractive and informative statistical graphics.

## About the Pokemon dataset(Kaggle) :

This data set includes 721 Pokemon, including their number, name, first and second type, and basic stats: HP, Attack, Defense, Special Attack, Special Defense, and Speed. It has been of great use when teaching statistics to kids. With certain types you can also give a geeky introduction to machine learning.

This are the raw attributes that are used for calculating how much damage an attack will do in the games. This dataset is about the pokemon games (NOT pokemon cards or Pokemon Go).

The data as described by Myles O'Neill is:

'#': ID for each pokemon

Name: Name of each pokemon

Type 1: Each pokemon has a type, this determines weakness/resistance to attacks

Type 2: Some pokemon are dual type and have 2

Total: sum of all stats that come after this, a general guide to how strong a pokemon is

HP: hit points, or health, defines how much damage a pokemon can withstand before fainting

Attack: the base modifier for normal attacks (eg. Scratch, Punch)

Defense: the base damage resistance against normal attacks

SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)

SP Def: the base damage resistance against special attacks

Speed: determines which pokemon attacks first each round

# A - [Line Plot](https://seaborn.pydata.org/generated/seaborn.lineplot.html?highlight=lineplot#seaborn.lineplot) 

### 1) hue
### 2) style
### 3) .set(rc={'figure.figsize':(20,10)})
### 4) ci = False
### 5) markers = True

```python
# Example 1: Simple one Line Graph
    
# Assuming we want to investigate the Attack vs Defence

ax = sns.lineplot(x = 'Attack', y = 'Defense', data = raw_data)
```

```python
#Example 2: Adding Categories

# huevector or key in data : Grouping variable that will produce lines with different colors.
# In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding.
# Can be either categorical or numeric, although color mapping will behave differently in latter case.

sns.set(rc={'figure.figsize':(20,10)})

ax = sns.lineplot(x = 'Attack', y = 'Defense', 
                  data = raw_data, hue = 'Generation')
```

```python
#Example 3: By generation, (change the style of all the 6 lines)

ax = sns.lineplot(x = 'Attack', y = 'Defense', 
                  data = raw_data, hue = 'Generation', style = 'Generation')
```

```python
#Example 4: By generation with style & increase the size

#increase the size

#set figure size by passing dictionary to rc parameter with key 'figure.figsize' in seaborn set method:

sns.set(rc={'figure.figsize':(20,10)})

ax = sns.lineplot(x = 'Attack', y = 'Defense', 
                  data = raw_data, hue = 'Generation', style = 'Generation')
```

```python
#remove ci(confidence interval lines to make it more clearer)

sns.set(rc={'figure.figsize':(20,10)})

ax = sns.lineplot(x = 'Attack', y = 'Defense', 
                  data = raw_data, hue = 'Generation', style = 'Generation', 
                  ci = False)
```

```python
#Example 5: Adding markers

sns.set(rc={'figure.figsize':(20,10)})

ax = sns.lineplot(x = 'Attack', y = 'Defense', 
                  data = raw_data, hue = 'Generation', style = 'Generation', 
                  ci = False, markers = True)
```

# B - [Bar Plot](https://seaborn.pydata.org/generated/seaborn.barplot.html)

#### 1) ci = False
#### 2) hue
#### 3) orient = 'h'
#### 4) color
#### 5) palette

Points to be remembered while creating barchart:

1) a bar plot shows only the mean (or other estimator) value.

2) the line signifies condfidence interval

```python
#Example 1: Total attack numbers as per Type 1

ax = sns.barplot(x = 'Type 1', y = 'Attack', data = raw_data)

# create a df with the mean values(to check the mean of all sub groups)
raw_data[['Attack', 'Type 1']].groupby('Type 1').agg({'Attack': 'sum'})
```

```python
#Example 2: remove confidence interval(ci)

#ax = sns.barplot(x = 'Type 1', y = 'Attack', data = raw_data, ci = None)
ax = sns.barplot(x = 'Type 1', y = 'Attack', data = raw_data, ci = False)
```

```python
#Example 3: remove ci and add hue

#ax = sns.barplot(x = 'Type 1', y = 'Attack', data = raw_data, ci = None)
ax = sns.barplot(x = 'Type 1', y = 'Attack', 
                 data = raw_data, ci = False, hue = 'Generation')
```

```python
#Example 4: changing direction from vertical to horizontal direction

ax = sns.barplot(x = 'Attack', y = 'Type 1', data = raw_data, 
                 ci = False, hue = 'Generation', orient = 'h')
```

```python
#Example 5: changing colors using color

ax = sns.barplot(x = 'Attack', y = 'Type 1', data = raw_data, 
                 ci = False, hue = 'Generation', orient = 'h', color = 'green')
```

```python
#changing colors using palette or we can change color using colorpicker

#ax = sns.barplot(x = x = 'Attack', y = 'Type 1', data = raw_data,ci = False, hue = 'Generation', orient = 'h', palette="#EC3F10")
ax = sns.barplot(x = 'Attack', y = 'Type 1', data = raw_data, 
                 ci = False, hue = 'Generation', orient = 'h', palette="Blues_d")
```

# C - [Histograms](https://seaborn.pydata.org/generated/seaborn.distplot.html)

#### 1) color
#### 2) distplot()
#### 3) with matplotlib's plt.axvline
#### 4) mean()
#### 5) hist = False

```python
# Example 1: Investigating the distribution for Attack numbers

x = raw_data['Attack'].values
sns.distplot(x, color = 'blue')
```

```python
#Example 2: Investigating the distribution of Attack, adding the mean using matplotlib(axvline)

#calculating the mean
mean_calc = raw_data['Attack'].mean()
mean_calc

sns.distplot(x, color = 'blue')

#plotting the mean using matplotlib
plt.axvline(mean_calc, 0,1, color = 'red')
```

```python
#Example 3: Investigating the distribution of revenue, adding the mean using matplotlib(axvline) and removing histogram

#hist = False 
x1 = raw_data['Defense'].values

#calculating the mean
mean_calc_1 = raw_data['Defense'].mean()
mean_calc_1 # 242.57440690150972

sns.distplot(x1, color = 'red', hist = False)

#plotting the mean using matplotlib
plt.axvline(mean_calc_1, 0,1, color = 'black')

```

# D - Box Plots

#### 1) boxplot()
#### 2) color
#### 3) hue
#### 4) boxplot() + swarmplot()

Important Note:

1) The middle line inside the box plot is median.

2) The box in the middle shows(starting point is Q1(25th percentile) and ending point is Q3(75th percentile))

3) The wiskers at both sides shows the minimum and maximum quartile.

4) The dots on the right are outliers.

```python
#Example 1: Investigating the distribution of Defense

x = raw_data['Defense'].values
x
sns.set(rc={'figure.figsize':(20,10)})

sns.boxplot(x)

print(raw_data['Defense'].median()) #70.0
```

```python
#Example 2: Investigating the distributionof Type 1 vs Defense

ax = sns.boxplot(x = 'Type 1', y = 'Defense', data = raw_data)
```

```python
#Example 3: Investigating the Defense vs Type 1  - horizontal(flip the sides) - change color

ax = sns.boxplot(x = 'Defense', y = 'Type 1', data = raw_data, color = 'blue')
```

```python
#Example 4: adding hue

ax = sns.boxplot(x = 'Type 2', y = 'Defense',data = raw_data, color = 'blue', hue = 'Generation')
```

```python
#Example 5: Adding datapoints with swarmplot(to add all data points)

ax = sns.boxplot(x = 'Type 2', y = 'Defense', data = raw_data, color = 'yellow')
ax = sns.swarmplot(x = 'Type 2', y = 'Defense',data = raw_data, color = 'red')
```
