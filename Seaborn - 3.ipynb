# G - SubPlots

### subplots method provides a way to plot multiple plots on a single figure. 
### Given the number of rows and columns , it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax .

#### 1) distplot()
#### 2) boxplot()
#### 3) swarmplot()
#### 4) matplotlib's subplots()

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
#Set the matplotlib figure

fig, axis = plt.subplots(2,2,figsize = (20,20))

a = raw_data['Attack'].values
b = raw_data['Defense'].values
c = raw_data['Speed'].values

#plot 1
sns.distplot(a, color = 'blue', ax = axis[0,0])

#plot 2
sns.distplot(b, color = 'blue', ax = axis[0,1])

#plot 3
sns.distplot(c, color = 'blue', ax = axis[1,0])

#plot 4

ax = sns.boxplot(x = 'Attack', y = 'Type 1',data = raw_data, color = 'yellow')
ax = sns.swarmplot(x = 'Attack', y = 'Type 1',data = raw_data, palette=pkmn_type_colors)
```

# H - [Pairplots](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

### seaborn.pairplot() :

### To plot multiple pairwise bivariate distributions in a dataset, we can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

#### 1) pairplot()
#### 2) plot_kws
#### 3) height
#### 4) raw_data[['col 1', 'col 2', col 3']] : running on specific columns
#### 5) hue 
#### 6) kind = reg

```python
# Example 1: running all dataframe - green color

g = sns.pairplot(raw_data, plot_kws = {'color': 'green'}, height = 6)
```

```python
raw_data.columns
```

```python
#Example 2: Running on specific columns

k = sns.pairplot(raw_data[['Attack', 'Defense', 'Speed']],
                 plot_kws = {'color' : 'green'}, height = 7)
```

```python
#Example 3: Adding hue

k = sns.pairplot(raw_data[['Attack', 'Defense', 'Speed', 'Type 1']], 
                 height = 5, hue = 'Type 1')
```

```python
raw_data.columns
```

```python
#Example 4: Adding hue and adding kind = reg

k = sns.pairplot(raw_data[['Attack', 'Defense', 'Speed', 'Generation']], 
                 height = 5, hue = 'Generation', kind = 'reg')
```

# I - [JointPlots](http://seaborn.pydata.org/generated/seaborn.jointplot.html)

### Seaborn's jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.This plot is a convenience class that wraps JointGrid.

#### 1) jointplot()
#### 2) color
#### 3) height
#### 4) kind = 'reg'

Drawing a plot using variables with bivariate and univariate graphs.

```python
#Example 1: Relationship between Attack , Defence column variables

ax = sns.jointplot('Attack','Defense', data = raw_data, color = 'green', height = 10, kind = 'reg')
```

# J - [Heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html)

### A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors.

#### 1) heatmap()
#### 2) corr()
#### 3) col
#### 4) annot
#### 5) annot_kws
#### 6) xticklabels
#### 7) yticklabels
#### 8) cmap

```python
#Example 1: Heatmap for numerical columns and calculate correlation between different variables

#calculate correlations
corr = raw_data[['Attack', 'Defense', 'Speed']].corr()

corr
```

```python
#Heatmap
sns.heatmap(corr)
```

```python
#Example 2: Using col, fig, annot, annot_kws, xticklabels and yticklabels 

#We can change the name of columns 
col = ['Forceful Attack ', 'Super Defence', 'Amazing Speed']

#using matplotlib fig
plt.figure(figsize=(12, 9))


#heatmap
sns.heatmap(corr,
            annot = True, 
            annot_kws = {'size': 10},
            xticklabels = col,
            yticklabels = col)
```

```python
#Example 3: using cmap (for different shades of colors)

col = ['Forceful Attack ', 'Super Defence', 'Amazing Speed']

#using matplotlib fig
plt.figure(figsize=(12, 9))


#heatmap
sns.heatmap(corr,
            annot = True, 
            annot_kws = {'size': 10},
            xticklabels = col,
            yticklabels = col,
            cmap = 'Blues')
```

```python
#Example 4: finding correlation between different column variables(Pearson correlation)

# Calculate correlations
corr_1 = raw_data.corr()
 
corr_1
```

```python
# Heatmap
plt.figure(figsize=(12, 9))

sns.heatmap(corr_1, annot = True, annot_kws = {'size': 10})
```
