```python
#Importing Packages

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

# E - [Scatter Plots](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)  

#### 1) scatterplot()
#### 2) color
#### 3) hue
#### 4) style
#### 5) size & sizes

```python
#Example 1: Relation between Attack and Defense 

ax = sns.scatterplot(x = 'Attack', y = 'Defense',data = raw_data, color = 'red')
```

```python
# Relationship between Attack and Defense , and changing color, hue on column 'Generation' and style on 'Generation'

ax = sns.scatterplot(x = 'Attack', y = 'Defense',data = raw_data, color = 'red', hue = 'Generation', style = 'Generation')
```

```python
# Example 3: changing color and hue -- adding size and sizes(according to sizes we can actually compare)

sns.set(rc={'figure.figsize':(20,10)})
ax = sns.scatterplot(x = 'Attack', y = 'Defense',data = raw_data, color = 'red', hue = 'Generation', style = 'Generation', size = 'Speed', sizes = (20,200))
```

# F - [lm Plots](https://seaborn.pydata.org/generated/seaborn.lmplot.html)  

This is basically a combination of Regplot() and Facetgrids

#### 1) lmplot()
#### 2) height
#### 2) hue
#### 3) ci
#### 4) markers
#### 5) col
#### 6) line_kws
#### 7) scatter_kws
#### 8) col_wrap

```python
#Example 1: lmplot() with height
    
ax = sns.lmplot(x = 'Attack', y = 'Defense', data = raw_data, height = 10)
```

```python
#Example 2: adding hue, ci and markers

ax = sns.lmplot(x = 'Attack', y = 'Defense', data = raw_data, height = 10, 
                hue = 'Generation', ci =None, markers = ['o', 'x', '^', '+', '*', '8'])
```

```python
#Example 3: Break down graphs into six groups by column

ax = sns.lmplot(x = 'Attack', y = 'Defense', data = raw_data, 
                height = 5, col = 'Generation', ci =None)
```

```python
#Example 4: Adding color to regression line and scatter plots on columns

ax = sns.lmplot(x = 'Attack', y = 'Defense', data = raw_data, 
                height = 5, col = 'Generation', ci =None,
                line_kws = {'color': 'red'},
                scatter_kws = {'color': 'green'})
```

```python
#Example 4: adding color_wrap

ax = sns.lmplot(x = 'Attack', y = 'Defense', data = raw_data, 
                height = 5, col = 'Generation', ci =None,
                line_kws = {'color': 'red'},
                scatter_kws = {'color': 'green'},
                col_wrap = 2)
```
